```python
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv("C:\\Users\\FADIYA\\Downloads\\Spotify Most Streamed Songs.csv")
df
```

Mencari Mean Median Modus

```python
#mengonversi kolom 'streams' menjadi numerik, dan mengabaikan error
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')

#Menghitung mean
mean_streams = df['streams'].mean()        # Rata-rata streams
median_streams = df['streams'].median()    # Median streams
mode_streams = df['streams'].mode()[0]     # Modus streams

print(f"Mean: {mean_streams}")
print(f"Median: {median_streams}")
print(f"Mode: {mode_streams}")
```

**1. Summary statistic dari data**

```python


# 1. Ukuran Pemusatan
mean_streams = df['streams'].mean()        # Rata-rata streams
median_streams = df['streams'].median()    # Median streams
mode_streams = df['streams'].mode()[0]     # Modus streams

# 2. Ukuran Penyebaran
range_streams = df['streams'].max() - df['streams'].min()  # Range streams
variance_streams = df['streams'].var()     # Varians streams
std_dev_streams = df['streams'].std()      # Standar Deviasi streams
iqr_streams = df['streams'].quantile(0.75) - df['streams'].quantile(0.25)  # IQR streams

# 3. Ukuran Posisi (untuk streams)
q1_streams = df['streams'].quantile(0.25)  # Kuartil pertama
q3_streams = df['streams'].quantile(0.75)  # Kuartil ketiga
min_streams = df['streams'].min()          # Minimum streams
max_streams = df['streams'].max()          # Maximum streams

# 4. Ukuran Bentuk Distribusi
skewness_streams = df['streams'].skew()    # Skewness (Kemiringan streams)
kurtosis_streams = df['streams'].kurt()    # Kurtosis streams


print("Ukuran Pemusatan:")
print(f"Mean (streams): {mean_streams}")
print(f"Median (streams): {median_streams}")
print(f"Mode (streams): {mode_streams}\n")

print("Ukuran Penyebaran:")
print(f"Range (streams): {range_streams}")
print(f"Variance (streams): {variance_streams}")
print(f"Standard Deviation (streams): {std_dev_streams}")
print(f"Interquartile Range (streams): {iqr_streams}\n")

print("Ukuran Posisi:")
print(f"Q1 (streams): {q1_streams}")
print(f"Q3 (streams): {q3_streams}")
print(f"Min (streams): {min_streams}")
print(f"Max (streams): {max_streams}\n")

print("Ukuran Bentuk Distribusi:")
print(f"Skewness (streams): {skewness_streams}")
print(f"Kurtosis (streams): {kurtosis_streams}")
```

Distribusi Tahun Rilis Lagu

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur ukuran grafik
plt.figure(figsize=(10,6))

# Membuat histogram dari released_year
sns.histplot(df['released_year'], bins=20, kde=False)

plt.title('Distribution of Song Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

Distribusi Jumlah Streams

```python
plt.figure(figsize=(10,6))

# Membuat histogram dari streams
sns.histplot(df['streams'], bins=30, kde=False, color='skyblue')

plt.title('Distribution of Song Streams')
plt.xlabel('Number of Streams')
plt.ylabel('Count')
plt.show()
```

Lagu Paling Banyak Diputar di Spotify dan Apple Playlists

```python
top_songs_spotify = df.nlargest(10, 'in_spotify_playlists')
top_songs_apple = df.nlargest(10, 'in_apple_playlists')

# Visualisasi top 10 lagu di Spotify Playlists
plt.figure(figsize=(10,6))
sns.barplot(data=top_songs_spotify, x='in_spotify_playlists', y='track_name', palette='viridis')
plt.title('Top 10 Songs in Spotify Playlists')
plt.xlabel('Number of Spotify Playlists')
plt.ylabel('Track Name')
plt.show()

# Visualisasi top 10 lagu di Apple Playlists
plt.figure(figsize=(10,6))
sns.barplot(data=top_songs_apple, x='in_apple_playlists', y='track_name', palette='magma')
plt.title('Top 10 Songs in Apple Playlists')
plt.xlabel('Number of Apple Playlists')
plt.ylabel('Track Name')
plt.show()
```

Distribusi Mode Lagu
Melihat distribusi mode lagu (Major atau Minor).

```python
plt.figure(figsize=(6,6))

# Membuat pie chart untuk distribusi mode
df['mode'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90, explode=(0.05, 0.05))

plt.title('Distribution of Song Modes (Major/Minor)')
plt.ylabel('')
plt.show()
```

Korelasi antar variabel numerik
Untuk melihat hubungan antara beberapa fitur numerik dalam dataset.

```python
plt.figure(figsize=(10,8))

# Membuat heatmap korelasi
correlation = df[['streams', 'danceability_%', 'energy_%', 'acousticness_%', 'valence_%']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')

plt.title('Correlation Heatmap')
plt.show()
```

Lagu-Lagu dengan Tingkat Instrumental Tinggi
Menampilkan 10 lagu dengan tingkat instrumentalness tertinggi.

```python
top_instrumental = df.nlargest(10, 'instrumentalness_%')

plt.figure(figsize=(10,6))

# Visualisasi top 10 lagu dengan instrumentalness tertinggi
sns.barplot(data=top_instrumental, x='instrumentalness_%', y='track_name', palette='cubehelix')

plt.title('Top 10 Songs with Highest Instrumentalness')
plt.xlabel('Instrumentalness (%)')
plt.ylabel('Track Name')
plt.show()
```

```python
# Menghapus baris yang mengandung NaN pada kolom 'streams' dan 'danceability_%'
df.dropna()

```

```python
# Mengisi NaN dengan rata-rata pada kolom 'streams' dan 'danceability_%'
df['streams'].fillna(df['streams'].mean(), inplace=True)
df['danceability_%'].fillna(df['danceability_%'].mean(), inplace=True)
df


```

**3. Lakukan uji korelasi**

```python
# Melakukan uji korelasi setelah mengisi NaN
correlation, p_value = stats.pearsonr(df['streams'], df['danceability_%'])

print(f"Korelasi Pearson antara Streams dan Danceability: {correlation}")
print(f"P-Value: {p_value}")
```
